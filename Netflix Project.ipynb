```python
num_movies = df[df.type == "MOVIE"].shape[0]
num_tv_shows = df[df.type == "SHOW"].shape[0]


labels = ["Movies", "TV Shows"]
counts = [num_movies, num_tv_shows]
plt.bar(labels, counts, color=['orange', 'purple'])
plt.title("Distribution of Movies and TV Shows on Netflix")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

```python
df_movies = df[df["type"] == "MOVIE"]
df_tv_shows = df[df["type"] == "SHOW"]


movie_release_years = df_movies["release_year"].value_counts().sort_index()
tv_show_release_years = df_tv_shows["release_year"].value_counts().sort_index()

plt.figure(figsize=(12, 6))
plt.plot(movie_release_years.index, movie_release_years.values, label="Movies")
plt.plot(tv_show_release_years.index, tv_show_release_years.values, label="TV Shows")
plt.xlabel("Release Year")
plt.ylabel("Count")
plt.title("Distribution of Movie and TV Show Release Years on Netflix")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.hist(df_movies["imdb_score"], bins=10, alpha=0.5, label="Movies")
plt.hist(df_tv_shows["imdb_score"], bins=10, alpha=0.5, label="TV Shows")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.title("Distribution of Movie and TV Show Ratings on Netflix")
plt.legend()
plt.show()
```

```python
movie_genres = df[df["type"] == "MOVIE"]["genres"].str.split(", ").explode().value_counts()
tv_show_genres = df[df["type"] == "SHOW"]["genres"].str.split(", ").explode().value_counts()

plt.figure(figsize=(12, 6))
plt.bar(movie_genres.index, movie_genres.values, label="Movies")
plt.bar(tv_show_genres.index, tv_show_genres.values, label="TV Shows")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.title("Most Common Genres of Movies and TV Shows on Netflix")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
num_movies = df[df["type"] == "MOVIE"]

plt.figure(figsize=(12, 6))
plt.scatter(df_movies["runtime"], df_movies["imdb_score"])
plt.xlabel("Duration (minutes)")
plt.ylabel("IMDB Rating")
plt.title("Correlation between Movie Duration and Rating on Netflix")
plt.show()
```

```python
df["runtime"] = df["runtime"].astype(str)

df["genres"] = df["genres"].str.split(", ")

df_duration = df[["genres", "runtime"]]

df_duration = df_duration.explode("genres")

df_duration["runtime"] = pd.to_numeric(df_duration["runtime"].str.replace(" min", ""), errors="coerce")

df_grouped = df_duration.groupby("genres").mean()

plt.figure(figsize=(12, 6))
plt.bar(df_grouped.index, df_grouped["runtime"])
plt.xlabel("Genre")
plt.ylabel("Duration (minutes)")
plt.title("Distribution of Movie and TV Show Durations by Genre on Netflix")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
```

```python
tv_shows = df[df.type == "SHOW"]
movies = df[df.type == "MOVIE"]

tv_shows_prop = len(tv_shows) / len(df) * 100
movies_prop = len(movies) / len(df) * 100

labels = ["TV Shows", "Movies"]
sizes = [tv_shows_prop, movies_prop]
colors = ["orange", "purple"]
plt.pie(sizes, labels=labels, colors=colors, autopct="%1.1f%%", startangle=90)
plt.title("Proportion of TV Shows and Movies on Netflix")
plt.axis("equal")
plt.show()
```

```python
age_ratings = movies["age_certification"].value_counts()

labels = age_ratings.index
sizes = age_ratings.values
colors = ["skyblue", "lightgreen", "orange", "pink", "yellow", "purple"]
plt.pie(sizes, labels=labels, colors=colors, autopct="%1.1f%%", startangle=180)
plt.title("Number of Movies with Different Age Ratings on Netflix")
plt.axis("equal")
plt.show()
```

```python
scores = movies[["imdb_score", "tmdb_score"]]

sns.heatmap(scores.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Between IMDB Score and TMDB Score for Movies on Netflix")
plt.show()
```
